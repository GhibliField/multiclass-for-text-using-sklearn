```python
import pandas as pd
trainset = pd.read_csv('trainset.csv',sep='\t')
devset = pd.read_csv('devset.csv',sep='\t')
en = pd.read_csv('seg_trainset_from_en.csv',sep='\t')
es = pd.read_csv('seg_trainset_from_es.csv',sep='\t')
ja = pd.read_csv('seg_trainset_from_ja.csv',sep='\t')
ko = pd.read_csv('seg_trainset_from_ko.csv',sep='\t')
fr = pd.read_csv('seg_trainset_from_fr.csv',sep='\t')


aug_concat=pd.concat([en,es,ja,ko,fr])

```

```python
label_cols = ["app","health","schedule","bus","lottery","stock","calc","map","telephone","match","train","cinemas","message","translation","contacts","music","tvchannel","cookbook","news","video","datetime","novel","weather","email","poetry","website","epg","radio","flight","riddle"]
aug_concat['chat'] = 1-aug_concat[label_cols].max(axis=1)#如果label_cols中不存在1(肯定是最大值),则chat标为１,否则标为0
small_aug=aug_concat.loc[(aug_concat['chat']!=1) 
              &(aug_concat['website']!=1)
              &(aug_concat['poetry']!=1)
              &(aug_concat['novel']!=1)
              &(aug_concat['video']!=1)
              &(aug_concat['music']!=1)
              &(aug_concat['stock']!=1)
              &(aug_concat['tvchannel']!=1)
              &(aug_concat['app']!=1)
              &(aug_concat['radio']!=1)
              &(aug_concat['telephone']!=1)]
```

```python
train_2018=pd.concat([trainset,devset,small_aug])
train_2018['chat'].fillna("0", inplace=True)
```

```python
train_2018.drop_duplicates(['text'])
train_2018.drop('id', axis=1, inplace=True)
train_2018.to_csv('trian_2018.csv',sep='\t',index=False)
```

```python
import os
filelist=os.listdir('./train')
filelist
```

```python
import jieba
import pandas as pd
import numpy as np
train_2017=pd.DataFrame()
for each in filelist:
    text=[]
    label=[]
    with open('./train/'+each,'r') as inp:
        for line in inp:
            text.append(' '.join([_ for _ in jieba.cut(line.strip())]))
            label.append('1')
    temp_=pd.DataFrame({'text':text})
    train_temp = pd.concat([temp_, pd.DataFrame(label, columns = [each.split('_')[-1].split('.')[0]])], axis=1)
    train_2017=pd.concat([train_2017,train_temp])
```

```python
train_2017
```

```python
filelist_dev=os.listdir('./develop')

dev_2017=pd.DataFrame()
for each in filelist_dev:
    text=[]
    label=[]
    with open('./develop/'+each,'r') as inp:
        for line in inp:
            text.append(' '.join([_ for _ in jieba.cut(line.strip())]))
            label.append('1')
    temp_dev=pd.DataFrame({'text':text})
    train_temp_dev = pd.concat([temp_dev, pd.DataFrame(label, columns = [each.split('_')[-1].split('.')[0]])], axis=1)
    dev_2017=pd.concat([dev_2017,train_temp_dev])
```

```python
data_2017=pd.concat([train_2017,dev_2017])
data_2017.drop_duplicates(['text'],inplace=True)
data_2017.fillna('0',inplace=True)
data_2017
```

```python
data_2017.to_csv('train_2017.csv',sep='\t')
tta_train=pd.concat([train_2018,data_2017])
tta_train.drop_duplicates(['text'],inplace=True)
tta_train.to_csv('tta_train.csv',sep='\t')
```

```python
filelist_dev=os.listdir('./test')

test_2017=pd.DataFrame()
for each in filelist_dev:
    text=[]
    label=[]
    with open('./test/'+each,'r') as inp:
        for line in inp:
            text.append(' '.join([_ for _ in jieba.cut(line.strip())]))
            label.append('1')
    temp_test=pd.DataFrame({'text':text})
    test_temp = pd.concat([temp_test, pd.DataFrame(label, columns = [each.split('_')[-1].split('.')[0]])], axis=1)
    test_2017=pd.concat([dev_2017,test_temp])

test_2017.drop_duplicates(['text'],inplace=True)
test_2017.fillna('0',inplace=True)
test_2017.to_csv('test.csv',sep='\t')
```
